```python
import pandas as pd
```

```python
df = pd.read_csv('../out/deka_criminal_cases_3.csv')
len(df)
```

```python
df.head()
```

```python
long_summs = df['long_text'].tolist()
short_summs = df['short_text'].tolist()
```

```python
def get_stats(text_list):
    count_no_text = 0
    for text in text_list:
        if text == '\"\"':
            count_no_text += 1
    return {
        'no_text_percentage' : count_no_text / len(df)
    }
```

```python
print(f'Percentage of cases with no long summary: {get_stats(long_summs)["no_text_percentage"] * 100:.3f} %')
print(f'Percentage of cases with no short summary: {get_stats(short_summs)["no_text_percentage"] * 100:.3f} %')
```
